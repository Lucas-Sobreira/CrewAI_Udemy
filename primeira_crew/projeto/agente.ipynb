### Import das Bibliotecas do CrewAI

```python
from crewai import Agent, Task, Crew, Process
```

### Criando meu Agente - Planejador de Viagens

#### Agents

```python
planejador_de_viagem = Agent(
    role="Planejador de Viagem", 
    goal="Planejar todos os detalhes de uma viagem, incluindo roteiros e atividades",
    backstory = """
        VocÃª Ã© um especialista em planejamento de viagens, sempre em busca de novas aventuras e experiÃªncias. 
        Seu objetivo Ã© garantir que os detalhes da viagem sejam organizados de maneira eficiente e agradÃ¡vel.
    """, 
    verbose=True
)
```

```python
orcamentista = Agent(
    role = "OrÃ§amentista de Viagem", 
    goal = "Estimar o custo total de uma viagem, considerando transporte, hospedagem, alimentaÃ§Ã£o e atividades",
    backstory = """
        VocÃª Ã© um analista financeiro focado em viagem. Sua missÃ£o Ã© garantir que os custos estejam dentro do orÃ§amento, 
        criando estimativas precisas para cada parte da viagem.
    """, 
    verbose=True
)
```

#### Tasks

```python
planeja_roteiro = Task(
    description = "Crie um roteiro detalhado para uma viagem, incluindo as cidades, atividades e transporte",
    agent=planejador_de_viagem,
    expected_output = """
        Roteiro com a sequÃªncia de cidades a serem visitadas, as principais atividades e o tipo de transporte utilizado.
    """
)
```

```python
estima_orcamento = Task(
    description = "Calcule o orÃ§amento total da viagem, levando em consideraÃ§Ã£o as cidades, transporte, hospedagem e atividades",
    agent=orcamentista,
    expected_output = """
        Estimativa de orÃ§amento detalhada, com custos aproximados para cada item da viagem
    """
)
```

#### Criando a Crew

```python
viagem_crew = Crew(
    agents = [planejador_de_viagem, orcamentista],
    tasks = [planeja_roteiro, estima_orcamento],
    process = Process.sequential
)
```

```python
result = viagem_crew.kickoff()
```
